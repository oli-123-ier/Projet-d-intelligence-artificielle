```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

# Step 1: Load and Prepare the Data
data_path = os.getcwd()  # Current project directory
labels_file = os.path.join(data_path, "train_labels.csv")
images_path = os.path.join(data_path, "images")

# Load labels
data = pd.read_csv(labels_file)
data["image_path"] = data["image"].apply(lambda x: os.path.join(images_path, x))
data['target'] = data['target'].astype(str)
# Check data
def print_data_info():
    print("Data Preview:")
    print(data.head())
    print("\nClass Distribution:")
    print(data['target'].value_counts())

print_data_info()

# Split the data
train_data, test_data = train_test_split(data, test_size=0.2, stratify=data['target'], random_state=42)

# Data Generators
data_gen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

train_generator = data_gen.flow_from_dataframe(
    dataframe=train_data,
    x_col="image_path",
    y_col="target",
    target_size=(128, 128),
    batch_size=32,
    class_mode="binary",
    subset="training"
)

validation_generator = data_gen.flow_from_dataframe(
    dataframe=train_data,
    x_col="image_path",
    y_col="target",
    target_size=(128, 128),
    batch_size=32,
    class_mode="binary",
    subset="validation"
)

# Test Data Generator
test_gen = ImageDataGenerator(rescale=1./255)
test_generator = test_gen.flow_from_dataframe(
    dataframe=test_data,
    x_col="image_path",
    y_col="target",
    target_size=(128, 128),
    batch_size=32,
    class_mode="binary",
    shuffle=False
)

# Step 2: Define the CNN Model
def create_model():
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = create_model()

# Step 3: Train the Model
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=10,
    verbose=1
)

# Step 4: Evaluate the Model
loss, accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Generate Classification Report
y_pred = (model.predict(test_generator) > 0.5).astype(int)
y_true = test_data['target'].values
y_true = y_true.astype(int)
print(classification_report(y_true, y_pred))

# Step 5: Plot Training History
def plot_training_history():
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.legend()
    plt.title('Model Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.legend()
    plt.title('Model Loss')
    plt.show()

plot_training_history()

# Step 6: Save the Model
model.save("mask_detection_model.h5")
print("Model saved as 'mask_detection_model.h5'")
```

```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

# Step 1: Load and Prepare the Data
data_path = os.getcwd()  # Current project directory
labels_file = os.path.join(data_path, "train_labels.csv")
images_path = os.path.join(data_path, "images")

# Load labels
data = pd.read_csv(labels_file)
data["image_path"] = data["image"].apply(lambda x: os.path.join(images_path, x))
data['target'] = data['target'].astype(str)


# data = pd.read_csv(labels_file)
# data["image_path"] = data["image"].apply(lambda x: os.path.join(images_path, x))
# data['target'] = data['target'].astype(str)
# Check data
def print_data_info():
    print("Data Preview:")
    print(data.head())
    print("\nClass Distribution:")
    print(data['target'].value_counts())

print_data_info()

# Split the data
train_data, test_data = train_test_split(data, test_size=0.2, stratify=data['target'], random_state=42)

# Data Generators
data_gen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    brightness_range=[0.8, 1.2],
    horizontal_flip=True
)

train_generator = data_gen.flow_from_dataframe(
    dataframe=train_data,
    x_col="image_path",
    y_col="target",
    target_size=(128, 128),
    batch_size=64,
    class_mode="binary",
    subset="training"
)

validation_generator = data_gen.flow_from_dataframe(
    dataframe=train_data,
    x_col="image_path",
    y_col="target",
    target_size=(128, 128),
    batch_size=64,
    class_mode="binary",
    subset="validation"
)

# Test Data Generator
test_gen = ImageDataGenerator(rescale=1./255)
test_generator = test_gen.flow_from_dataframe(
    dataframe=test_data,
    x_col="image_path",
    y_col="target",
    target_size=(128, 128),
    batch_size=64,
    class_mode="binary",
    shuffle=False
)

# Step 2: Define the Improved CNN Model
def create_model():
    model = Sequential([
        Conv2D(64, (3, 3), activation='relu', input_shape=(128, 128, 3)),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.3),

        Conv2D(128, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.3),

        Conv2D(256, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.4),

        Flatten(),
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005), loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = create_model()

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
model_checkpoint = ModelCheckpoint("best_model.h5", save_best_only=True, monitor='val_loss')

# Step 3: Train the Model
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=30,
    # callbacks=[early_stop, reduce_lr, model_checkpoint],
    verbose=1
)

# Step 4: Evaluate the Model
loss, accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Generate Classification Report
y_pred = (model.predict(test_generator) > 0.5).astype(int)
y_true = test_data['target'].values
y_true = y_true.astype(int)
print(classification_report(y_true, y_pred))

# Step 5: Plot Training History
def plot_training_history():
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.legend()
    plt.title('Model Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.legend()
    plt.title('Model Loss')
    plt.show()

plot_training_history()

# Step 6: Save the Best Model
model.save("mask_detection_best_model.h5")
print("Model saved as 'mask_detection_best_model.h5'")
print("Best model saved as 'best_model.h5'")
```

```python
import tensorflow as tf
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import img_to_array

# Charger le modèle entraîné
# model = tf.keras.models.load_model("/content/best_model.h5")
model = tf.keras.models.load_model("/content/mask_detection_model.h5")
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Chemin du dossier contenant les images à tester
test_images_path = "/content/test_img"  # Modifier selon l'emplacement des images

# Classes
classes = ["Sans Masque", "Avec Masque"]

def preprocess_image(image_path):
    """Charge et prétraite une image pour la prédiction."""
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (128, 128))  # Modifier selon la taille d'entrée du modèle
    image = img_to_array(image) / 255.0  # Normalisation
    image = np.expand_dims(image, axis=0)  # Ajouter une dimension batch
    return image

def predict_image(image_path):
    """Effectue une prédiction sur une image donnée."""
    image = preprocess_image(image_path)
    prediction = model.predict(image)[0][0]
    class_label = classes[1] if prediction > 0.5 else classes[0]
    return class_label, prediction

# Tester sur quelques images
test_images = [os.path.join(test_images_path, img) for img in os.listdir(test_images_path) if img.endswith(".jpg") or img.endswith(".png")]

for image_path in test_images:
    label, confidence = predict_image(image_path)
    print(f"Image: {image_path}, Prédiction: {label} ({confidence:.2f})")

    # Affichage de l'image avec la prédiction
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(f"{label} ({confidence:.2f})")
    plt.axis("off")
    plt.show()
```

```python
# Installer les dépendances nécessaires (si elles ne sont pas déjà installées)
!pip install tensorflow pandas numpy matplotlib

import os
import shutil
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Configuration des chemins
from google.colab import drive
drive.mount('/content/drive')

BASE_DIR = '/content/drive/My Drive/your_folder_name'  # Remplacez par le chemin de votre dossier dans Google Drive
DATA_DIR = os.path.join(BASE_DIR, 'archive')
IMAGE_DIR = os.path.join(DATA_DIR, 'images')

TRAIN_DIR = os.path.join(IMAGE_DIR, 'train')
TEST_DIR = os.path.join(IMAGE_DIR, 'valid')

TRAIN_CSV = os.path.join(DATA_DIR, 'train_labels.csv')
TEST_CSV = os.path.join(DATA_DIR, 'sample_sub.csv')

print(TRAIN_DIR)
print(TRAIN_CSV)
print(TEST_DIR)
print(TEST_CSV)

# Validation des fichiers
if not os.path.exists(TRAIN_CSV) or not os.path.exists(TEST_CSV):
    raise FileNotFoundError("Les fichiers CSV nécessaires sont introuvables.")

# Chargement des labels
train_labels = pd.read_csv(TRAIN_CSV)

# Création des sous-dossiers pour l'entraînement
for label in ['withMask', 'noMask']:
    label_dir = os.path.join(TRAIN_DIR, label)
    os.makedirs(label_dir, exist_ok=True)

# Organisation des images
for _, row in train_labels.iterrows():
    img_name = row['image']
    label = 'withMask' if row['target'] == 1 else 'noMask'
    src_path = os.path.join(TRAIN_DIR, img_name)
    dst_path = os.path.join(TRAIN_DIR, label, img_name)

    if os.path.exists(src_path):
        shutil.move(src_path, dst_path)
    else:
        print(f"Image non trouvée : {src_path}")

# Génération des ensembles de données
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

# Chargement du modèle préentraîné
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(150, 150, 3))

# Ajout des couches personnalisées
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
final_output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=final_output)

# Geler les couches du modèle préentraîné
for layer in base_model.layers:
    layer.trainable = False

# Compilation du modèle
model.compile(optimizer=Adam(learning_rate=0.00001), loss='binary_crossentropy', metrics=['accuracy'])

# Configuration des callbacks
callbacks = [ EarlyStopping(patience=5, restore_best_weights=True), ReduceLROnPlateau(factor=0.1, patience=3)]

# Entraînement du modèle
history = model.fit( train_generator, validation_data=val_generator, epochs=30, callbacks=callbacks )

# Visualisation des résultats
def plot_training_history(history):
    plt.figure(figsize=(12, 4))

    # Courbe de précision
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    # Courbe de perte
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()

plot_training_history(history)

# Sauvegarde du modèle
model.save(os.path.join(BASE_DIR, 'inception_v3_mask_detector.h5'))

print("Modèle entraîné et sauvegardé avec succès.")
```
