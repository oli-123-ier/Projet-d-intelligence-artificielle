```python
# prompt: meme chose pour le filtre canny (pour les contour)

import cv2
import matplotlib.pyplot as plt
import os

dossier_images = "./IMAGES"  # Spécifiez le bon chemin du dossier
chemins_images = []

# Parcours du dossier d'images et ajout des chemins à la liste
for nom_fichier in os.listdir(dossier_images):
  chemin_image = os.path.join(dossier_images, nom_fichier)
  if os.path.isfile(chemin_image):
    chemins_images.append(chemin_image)

# Parcours des images et application du filtre Canny
for chemin_image in chemins_images:
  try:
    # Lecture de l'image en niveaux de gris
    img = cv2.imread(chemin_image, cv2.IMREAD_GRAYSCALE)

    # Application du filtre Canny avec différents seuils
    edges_50_100 = cv2.Canny(img, 50, 100)
    edges_100_200 = cv2.Canny(img, 100, 200)


    # Affichage des images
    plt.figure(figsize=(15, 5))
    plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title('Original')
    plt.subplot(132), plt.imshow(edges_50_100, cmap='gray'), plt.title('Canny (50, 100)')
    plt.subplot(133), plt.imshow(edges_100_200, cmap='gray'), plt.title('Canny (100, 200)')
    plt.show()

  except Exception as e:
    print(f"Erreur lors du traitement de l'image {chemin_image}: {e}")
```

```python
# prompt: pour le filtre souber

import cv2
import matplotlib.pyplot as plt
import os

dossier_images = "./IMAGES"  # Spécifiez le bon chemin du dossier
chemins_images = []

# Parcours du dossier d'images et ajout des chemins à la liste
for nom_fichier in os.listdir(dossier_images):
  chemin_image = os.path.join(dossier_images, nom_fichier)
  if os.path.isfile(chemin_image):
    chemins_images.append(chemin_image)

# Parcours des images et application du filtre Sobel
for chemin_image in chemins_images:
  try:
    # Lecture de l'image en niveaux de gris
    img = cv2.imread(chemin_image, cv2.IMREAD_GRAYSCALE)

    # Application du filtre Sobel
    sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
    sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
    abs_grad_x = cv2.convertScaleAbs(sobelx)
    abs_grad_y = cv2.convertScaleAbs(sobely)
    grad = cv2.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)

    # Affichage des images
    plt.figure(figsize=(15, 5))
    plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title('Original')
    plt.subplot(132), plt.imshow(abs_grad_x, cmap='gray'), plt.title('Sobel X')
    plt.subplot(133), plt.imshow(grad, cmap='gray'), plt.title('Sobel Combined')
    plt.show()

  except Exception as e:
    print(f"Erreur lors du traitement de l'image {chemin_image}: {e}")
```
