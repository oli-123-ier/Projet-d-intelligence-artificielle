```python
# Installation
!pip install ultralytics opencv-python matplotlib gdown

# Import necessary libraries
from IPython import get_ipython
from IPython.display import display
import gdown
import cv2
import matplotlib.pyplot as plt
import os
import random
import torch
from ultralytics import YOLO
import glob
import shutil #Added to move files


# Download the dataset
id_copied_on_google_drive = "1qDtvcFKHD64TS2f_Q-Sj8-MQQnegxpGd"
url = f"https://drive.google.com/uc?id={id_copied_on_google_drive}"
out = "Construction_Site_Safety_Dataset.zip"
gdown.download(url, out, quiet=False)
!unzip Construction_Site_Safety_Dataset.zip

# --- Data exploration and preparation ---
# Chemin vers le dossier des images
image_dir = "/content/Construction_Site_Safety_Dataset/images/train/images"
# Sélectionner 5 images aléatoires
sample_images = random.sample(os.listdir(image_dir), 5)

# Affichage des images
plt.figure(figsize=(15, 5))
for i, img_name in enumerate(sample_images):
    img_path = os.path.join(image_dir, img_name)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(1, 5, i+1)
    plt.imshow(img)
    plt.title(img_name)
    plt.axis("off")

plt.show()

# Chemin vers le dossier des annotations
label_dir = "/content/Construction_Site_Safety_Dataset/labels/train/labels"

# Sélectionner une image et son annotation
sample_img = sample_images[0]
img_path = os.path.join(image_dir, sample_img)
label_path = os.path.join(label_dir, sample_img.replace(".jpg", ".txt"))

# Charger et afficher l'annotation
with open(label_path, "r") as f:
    annotations = f.readlines()

print(f"Annotations de {sample_img} :\n")
for ann in annotations:
    print(ann.strip())  # Afficher chaque ligne de l’annotation

# Redimensionnement
output_dir = "/content/Construction_Site_Safety_Dataset/images_resized/train"
os.makedirs(output_dir, exist_ok=True)

for img_name in os.listdir(image_dir):
    img_path = os.path.join(image_dir, img_name)
    img = cv2.imread(img_path)

    # Redimensionnement
    resized_img = cv2.resize(img, (640, 640))

    # Sauvegarde
    cv2.imwrite(os.path.join(output_dir, img_name), resized_img)

print("✅ Redimensionnement terminé !")

# Vérifier si un GPU est disponible
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🔍 Utilisation de : {device}")

# Correction des annotations
label_dir = "/content/Construction_Site_Safety_Dataset/labels/train/labels"

for label_file in glob.glob(os.path.join(label_dir, "*.txt")):
    with open(label_file, "r") as f:
        lines = f.readlines()

    new_lines = []
    for line in lines:
        parts = line.strip().split()
        if len(parts) == 5:  # Format correct
            new_lines.append(line)
        elif len(parts) == 4:  # Format incorrect → ajouter une valeur par défaut
            print(f"⚠️ Annotation incomplète dans {label_file}: {line.strip()}")
            parts.append("0.1")  # Ajout d'une largeur/hauteur par défaut
            new_lines.append(" ".join(parts) + "\n")

    # Sauvegarder les annotations corrigées
    with open(label_file, "w") as f:
        f.writelines(new_lines)

print("✅ Toutes les annotations ont été vérifiées et corrigées !")

# Nettoyage des données
image_dir = "/content/Construction_Site_Safety_Dataset/images/train/images"
label_dir = "/content/Construction_Site_Safety_Dataset/labels/train/labels"

# Lister les fichiers
image_files = {f.split('.')[0] for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png'))}
label_files = {f.split('.')[0] for f in os.listdir(label_dir) if f.endswith('.txt')}

# Vérifier les fichiers qui posent problème
images_sans_labels = image_files - label_files
labels_sans_images = label_files - image_files

print(f"🟡 Images sans labels : {len(images_sans_labels)}")
print(f"🟡 Labels sans images : {len(labels_sans_images)}")

# Supprimer les fichiers problématiques
for img in images_sans_labels:
    img_path = os.path.join(image_dir, img + ".jpg")  # Modifier selon l'extension
    if os.path.exists(img_path):
        os.remove(img_path)
        print(f"❌ Image supprimée : {img_path}")

for lbl in labels_sans_images:
    lbl_path = os.path.join(label_dir, lbl + ".txt")
    if os.path.exists(lbl_path):
        os.remove(lbl_path)
        print(f"❌ Label supprimé : {lbl_path}")

print("✅ Nettoyage des fichiers orphelins terminé !")

# Vérification des données
train_images = "/content/Construction_Site_Safety_Dataset/images/train/images"
train_labels = "/content/Construction_Site_Safety_Dataset/labels/train/labels"

# Vérifier le nombre de fichiers restants
num_images = len(os.listdir(train_images))
num_labels = len(os.listdir(train_labels))

print(f"📂 Images restantes : {num_images}")
print(f"📂 Labels restants : {num_labels}")

if num_images == 0 or num_labels == 0:
    print("❌ Erreur : Aucune donnée disponible pour l'entraînement !")

# --- Création du jeu de validation ---
# Create validation folders if they don't exist
val_images_dir = "/content/Construction_Site_Safety_Dataset/images/valid/images"
val_labels_dir = "/content/Construction_Site_Safety_Dataset/labels/valid/labels"

os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# Move a portion of data from training to validation set (e.g., 10%)
all_images = os.listdir(train_images)
num_val_images = int(len(all_images) * 0.1) #10% of images will be used for validation
val_images = random.sample(all_images, num_val_images)

for img_name in val_images:
    # Moving images
    img_src_path = os.path.join(train_images, img_name)
    img_dest_path = os.path.join(val_images_dir, img_name)
    shutil.move(img_src_path, img_dest_path)
    print(f"Image déplacée: {img_name} vers {val_images_dir}")
    # Moving labels
    lbl_name = img_name.replace(".jpg", ".txt")  # Assuming labels have .txt extension
    lbl_src_path = os.path.join(train_labels, lbl_name)
    lbl_dest_path = os.path.join(val_labels_dir, lbl_name)
    if os.path.exists(lbl_src_path):
        shutil.move(lbl_src_path, lbl_dest_path)
        print(f"Label déplacé: {lbl_name} vers {val_labels_dir}")

print("✅ Jeu de validation créé et peuplé !")
# --- Mise à jour du fichier data.yaml ---
yaml_content = """path: /content/Construction_Site_Safety_Dataset
train: images/train/images
val: images/valid/images
test: images/test/images

nc: 3
names: ['Helmet', 'Vest', 'No_PPE']
"""

with open("/content/Construction_Site_Safety_Dataset/data.yaml", "w") as file:
    file.write(yaml_content)

print("✅ Fichier data.yaml mis à jour avec succès !")

# Print a label content
label_path = "/content/Construction_Site_Safety_Dataset/labels/train/labels/IMG_5024_mp4-4_jpg.rf.20e82f90e1a93960b6d3ba20323ee4d5.txt"

with open(label_path, "r") as f:
    print(f.readlines())

# --- Lancer l'entraînement ---
# Initialize the model
model = YOLO("yolov8n.pt")
# Train the model
model.train(
    data="/content/Construction_Site_Safety_Dataset/data.yaml",
    epochs=50,
    batch=16,
    imgsz=640,
    device=device,
    patience=5,
    optimizer="Adam",
    lr0=0.001,
    weight_decay=0.0005,
    augment=True
)
```
