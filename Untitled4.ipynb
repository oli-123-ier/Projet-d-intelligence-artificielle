```python
import pandas as pd
import numpy as np

# Chemin du fichier
file_path = "/content/aws_faq.csv"

# Charger le dataset
df = pd.read_csv(file_path)

# Afficher les premières lignes pour voir la structure
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Vectorisation des questions avec TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df["Question"])

def find_best_answer(user_question):
    """Trouve la question la plus proche et retourne la réponse."""
    user_tfidf = vectorizer.transform([user_question])  # Transformer la question de l'utilisateur
    similarities = cosine_similarity(user_tfidf, tfidf_matrix)  # Calculer la similarité cosinus

    best_match_idx = similarities.argmax()  # Indice de la question la plus similaire
    best_question = df.iloc[best_match_idx]["Question"]
    best_answer = df.iloc[best_match_idx]["Answer"]

    return best_question, best_answer

# Test rapide
test_question = "What is EC2?"
find_best_answer(test_question)
```

```python
# Remplacer les valeurs NaN par des chaînes vides
df = df.fillna("")

# Recalculer la matrice TF-IDF après le nettoyage
tfidf_matrix = vectorizer.fit_transform(df["Question"])

# Refaire le test
find_best_answer(test_question)
```
