```python
import glob
import cv2
import matplotlib.pyplot as plt
import os
import gdown


# images = glob.glob("A")
# masks = glob.glob("Abdomen_Only_Dataset/Dataset8_Abdomen/train/skin_masks/*.png")
dossier_images = "./IMAGES"  # Spécifiez le bon chemin du dossier
chemins_images = []

print(len(dossier_images))
# Parcours du dossier d'images et ajout des chemins à la liste
for nom_fichier in os.listdir(dossier_images):
  chemin_image = os.path.join(dossier_images, nom_fichier)
  if os.path.isfile(chemin_image):
    chemins_images.append(chemin_image)

# Parcours des images et affichage
for chemin_image in chemins_images:
  try:
    # Lecture de l'image avec OpenCV
    img = cv2.imread(chemin_image)

    # Conversion de l'image de BGR à RGB pour Matplotlib
    # img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # cv2.blur()
    plt.imshow(img, cmap=plt.cm.gray)
    plt.show()



  except Exception as e:
    print(f"Erreur lors du traitement de l'image {chemin_image}: {e}")

```
