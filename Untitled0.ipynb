```python
import cv2
from matplotlib import pyplot as plt
from scipy import misc

# Lecture de l'image
image = cv2.imread('/content/photo2.jpg', cv2.IMREAD_GRAYSCALE)



# Affichage de l'image
plt.imshow(image, cmap=plt.cm.gray)
plt.show()
```

```python
# Calcul de l'histogramme avec OpenCV
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Affichage de l'histogramme
plt.plot(hist)
plt.title('Histogramme avec OpenCV')
plt.xlabel('Intensit des pixels')
plt.ylabel('Nombre de pixels')
plt.show()
```

```python
import numpy as np

# Fonction pour calculer l'histogramme manuellement
def calculate_histogram(image):
    histogram = np.zeros(256, dtype=int)
    for pixel in image.ravel():
        histogram[pixel] += 1
    return histogram

# Calcul de l'histogramme
manual_hist = calculate_histogram(image)

# Affichage de l'histogramme
plt.plot(manual_hist)
plt.title('Histogramme calcul manuellement')
plt.xlabel('Intensit des pixels')
plt.ylabel('Nombre de pixels')
plt.show()
```
